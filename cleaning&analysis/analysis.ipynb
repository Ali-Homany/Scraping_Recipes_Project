```python
import pandas as pd
df_cleaning=pd.read_json('cleaning_data.json')
cuisine_ratings = df_cleaning.groupby('cuisine')['rating'].mean()

# Find the cuisine with the highest average rating
favorite_cuisine = cuisine_ratings.idxmax()

# Get the highest average rating
highest_rating = cuisine_ratings.max()

print("Favorite Cuisine:", favorite_cuisine)
print("Average Rating:", highest_rating)
```

```python

df_cleaning['num_elements'] = df_cleaning['steps'].apply(lambda x: len(x))
correlation = df_cleaning['num_elements'].corr(df_cleaning['rating'])
correlation
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_cleaning['num_elements'], df_cleaning['rating'])
plt.xlabel('num_elements')
plt.ylabel('Rating')
plt.title('num_elements vs. Rating')
plt.show()
```

```python
difficult_of_cuisine=df_cleaning.groupby('cuisine')['num_elements'].mean()

most_difficult_cuisine = difficult_of_cuisine.idxmax()
highest_difficult_avg = difficult_of_cuisine.max()
```
